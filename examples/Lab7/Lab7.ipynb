## add_image()

```python
m = gu.Map()

glaciernp_bounds = ((48.40, -114.15), (48.95, -113.40))

url = "https://upload.wikimedia.org/wikipedia/commons/9/97/Logo_of_the_United_States_National_Park_Service.svg?utm_source=commons.wikimedia.org&utm_campaign=index&utm_content=original"

image = m.add_image(url, glaciernp_bounds, opacity=1, name="image")

m.add_layer_control()

m.fit_bounds(glaciernp_bounds)

m
```

## add_video()

```python
m = gu.Map()

bounds = [[13.0, -130.0], [32.0, -100.0]]

url = "https://www.mapbox.com/bites/00188/patricia_nasa.mp4"

m.add_video(url, bounds, opacity=1)

m.add_layer_control()
m.fit_bounds(bounds)

m
```

## add_wms_layer()

```python
m = gu.Map(center=(38.51378825951165, -92.27416992187501), zoom=7)

url = "https://stateimagery.msdis.missouri.edu/arcgis/services/Missouri_6inch_Statewide_2023_2024_Dynamic/ImageServer/WMSServer"

layers = "Missouri_6inch_Statewide_2023_2024_Dynamic"

m.add_wms_layer(url, layers)

m
```
