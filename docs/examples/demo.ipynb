## Using folium implementation

```python
import geog510_lab_utils.foliumap as gu
```

```python
m = gu.Map(center=(40, -100), zoom=4, size="400px")

data_url = "https://github.com/opengeos/datasets/releases/download/us/us_states.geojson"

m.add_vector(data_url, name="US States")

m.add_layer_control()

m
```

## Using ipyleaflet implementation

```python
import geog510_lab_utils.geog510_lab_utils as gu
```

```python
m = gu.Map(center=(40, -100), zoom=4, size="400px")

data_url = "https://github.com/opengeos/datasets/releases/download/us/us_states.geojson"

m.add_basemap()

m.add_vector(data_url, name="US States")

m.add_layer_control()

m
```
